```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PATH = 'D:\Federated learning\Dataset\Obfuscated-MalMem2022.csv'
data = pd.read_csv(PATH)
data.describe()
```

```python
# iloc là 1 phương thức của dataframe cho phép truy cập dữ liệu bằng chỉ số nguyên dựa trên vị trí
X=data.iloc[:, 1:]
y=data ['Category']
```

```python
data.iloc[:, 56:57]
```

```python
# Tách label từ cột category và lưu thành cột mới
labels=[]
for i in y:
  labels.append( i.split('-')[0])
data ['label'] = labels
#Đếm label lưu thành mảng
labels_counts = []
for i in range(0,4):
  labels_counts.append(data ['label'].value_counts()[i])
#Lưu mảng labels
labels_name = ['Benign', 'Spyware', 'Trojan', 'Ransomeware']
```

```python
plt.pie(labels_counts, labels = labels_name, autopct='%1.1f%%')
plt.title('Phân bố dữ liệu')
plt.show()
```

```python
plt.title('Biểu đồ phân bố các nhãn trích xuất')
plt.xlabel('label')
plt.ylabel('count')
plt.bar(labels_name, labels_counts)
print(data['label'].value_counts())
```
